<!-- Estudo - Top 5 Estados com Maior Populacao dos Estados Unidos -->

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Ler CSV com pandas
url = "https://raw.githubusercontent.com/plotly/datasets/master/2014_usa_states.csv"
df = pd.read_csv(url)
```

```python
print(df.head())
```

<!-- Pandas -->

```python
df = df.rename(columns={"State":"Estado", "Postal":"UF", "Population":"Populacao"})

print(df.head())
```

```python
top5 = df.head(5)

# Grafico de Barras com Matplotlib 
plt.bar(top5["UF"], top5["Populacao"], color='orange')
plt.title("Populacao por Estado")
plt.xlabel("UF")
plt.ylabel("Populacao")

for i, valor in enumerate(top5["Populacao"]):
    valor_formatado = f"{int(valor):,}".replace(",", ".")
    plt.text(i, valor + 1000, valor_formatado, ha='center', va='bottom')

plt.show()
```
